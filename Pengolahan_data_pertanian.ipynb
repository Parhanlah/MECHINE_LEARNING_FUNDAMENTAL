#**Melihat Selisih Luas Tanah Dengan Hasi Panen**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset menggunakan Pandas
dataset = pd.read_csv('data - Sheet1.csv')

# Pilih kolom 'luas_tanah' dan 'hasil_panen'
hasil_panen = dataset['hasil_panen']
luas_tanah = dataset['luas_tanah']

# Hitung rata-rata untuk setiap kolom
rata_rata_hasil_panen = hasil_panen.mean()
rata_rata_luas_tanah = luas_tanah.mean()

# Hitung selisih rata-rata
selisih_rata_rata =  rata_rata_hasil_panen - rata_rata_luas_tanah

# Tampilkan hasil perbandingan
print("Perbandingan antara luas_tanah dan hasil_panen:")
print(f"Rata-rata hasil_panen: {rata_rata_hasil_panen}")
print(f"Rata-rata luas_tanah: {rata_rata_luas_tanah}")
print(f"Selisih rata-rata: {selisih_rata_rata}")

# Visualisasi data
plt.figure(figsize=(12, 5))
plt.scatter(luas_tanah, hasil_panen, alpha=0.5)
plt.title('Perbandingan antara Luas Tanah dan Hasil Panen')
plt.xlabel('Luas Tanah')
plt.ylabel('Hasil Panen')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset menggunakan Pandas
dataset = pd.read_csv('data - Sheet1.csv')

# Pilih kolom 'jenis_padi', 'harga_bibit', dan 'hasil_panen'
jenis_padi = dataset['jenis_padi']
harga_bibit = dataset['harga_bibit']
hasil_panen = dataset['hasil_panen']

# Gabungkan kolom-kolom ke dalam satu DataFrame baru
rekomendasi_df = pd.DataFrame({'jenis_padi': jenis_padi, 'harga_bibit': harga_bibit, 'hasil_panen': hasil_panen})

# Urutkan DataFrame berdasarkan hasil panen terbanyak dan harga terendah
rekomendasi_df = rekomendasi_df.sort_values(by=['hasil_panen', 'harga_bibit'], ascending=[False, True])

# Tampilkan hasil rekomendasi
print("\nRekomendasi Harga Bibit berdasarkan Hasil Panen Terbanyak dan Harga Terendah:")
print(rekomendasi_df[['jenis_padi', 'hasil_panen', 'harga_bibit']].head())

# Visualisasi hasil rekomendasi dalam bentuk bar plot
plt.figure(figsize=(12, 6))
plt.bar(rekomendasi_df.index, rekomendasi_df['harga_bibit'], color='skyblue')
plt.title('Rekomendasi Harga Bibit berdasarkan Hasil Panen Terbanyak dan Harga Terendah')
plt.xlabel('Varietas Padi')
plt.ylabel('Harga Bibit')
plt.xticks(rekomendasi_df.index, rekomendasi_df['jenis_padi'], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#**Jenis Padi Yang Kami Rekomendasikan**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file CSV
data = pd.read_csv('data - Sheet1.csv')

# Tampilkan beberapa baris pertama data untuk melihat strukturnya
print("Beberapa Baris Pertama Data:")
print(data.head())

# Menghitung rata-rata hasil_panen per 1 hektar untuk setiap jenis padi
data['hasil_panen_per_hektar'] = data['hasil_panen'] / data['luas_tanah']

rata_rata_hasil_panen_per_hektar = data.groupby('jenis_padi')['hasil_panen_per_hektar'].mean()

# Rekomendasikan jenis padi berdasarkan rata-rata hasil panen tertinggi per 1 hektar
varietas_rekomendasi = rata_rata_hasil_panen_per_hektar.sort_values(ascending=False).index.tolist()

# Tampilkan hasil rekomendasi
print("\nJenis Padi yang Direkomendasikan Berdasarkan Rata-rata Hasil Panen Tertinggi per 1 Hektar:")
print(varietas_rekomendasi[:10])  # Menampilkan tiga jenis padi pertama

# Visualisasi hasil rekomendasi
plt.figure(figsize=(12, 3))
plt.bar(varietas_rekomendasi[:10], rata_rata_hasil_panen_per_hektar[varietas_rekomendasi[:10]])
plt.title('Rekomendasi Jenis Padi Berdasarkan Rata-rata Hasil Panen Tertinggi per 1 Hektar')
plt.xlabel('Jenis Padi')
plt.ylabel('Rata-rata Hasil Panen per 1 Hektar')
plt.show()
```

### **Rekomendasi harga padi**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file CSV
data = pd.read_csv('data - Sheet1.csv')

# Tampilkan beberapa baris pertama data untuk melihat strukturnya
print("Beberapa Baris Pertama Data:")
print(data.head())

# Analisis data, misalnya, menghitung rata-rata harga bibit untuk setiap varietas
rata_rata_harga_bibit = data.groupby('jenis_padi')['harga_bibit'].mean()

# Rekomendasikan harga bibit berdasarkan rata-rata harga tertinggi
harga_rekomendasi = rata_rata_harga_bibit.sort_values(ascending=False).to_dict()

# Tampilkan hasil rekomendasi
print("\nRekomendasi Harga Bibit berdasarkan Rata-rata Harga Bibit Tertinggi:")
for varietas, harga in harga_rekomendasi.items():
    print(f"Varietas: {varietas}, Harga Rekomendasi: {harga}")

# Visualisasi hasil rekomendasi
plt.figure(figsize=(12, 6))
plt.bar(harga_rekomendasi.keys(), harga_rekomendasi.values(), color='skyblue')
plt.title('Rekomendasi Harga Bibit berdasarkan Rata-rata Harga Bibit Tertinggi')
plt.xlabel('Varietas Padi')
plt.ylabel('Harga Rekomendasi Bibit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file CSV
data = pd.read_csv('data - Sheet1.csv')

# Tampilkan beberapa baris pertama data untuk melihat strukturnya
print("Beberapa Baris Pertama Data:")
print(data.head())

# Analisis data, menghitung rata-rata hasil panen untuk setiap varietas
rata_rata_hasil_panen = data.groupby('jenis_padi')['hasil_panen'].mean()

# Rekomendasikan bibit berdasarkan hasil panen terbanyak
padi_padi = rata_rata_hasil_panen.sort_values(ascending=False).to_dict()

# Tampilkan hasil rekomendasi
print("\nJenis Padi Berdasarkan Harga")
for varietas, jumlah in padi_padi.items():
    print(f"Varietas: {varietas}, Ton: {jumlah}")

# Visualisasi hasil rekomendasi
plt.figure(figsize=(12, 6))
plt.bar(padi_padi.keys(), padi_padi.values(), color='lightgreen')
plt.title('Jenis Padi Berdasarkan Jumlah Panen')
plt.xlabel('Varietas Padi')
plt.ylabel('Hasil Panen Dalam Ton')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#**Harga Bibit berdasarkan Hasil Panen**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset menggunakan Pandas
dataset = pd.read_csv('data - Sheet1.csv')

# Pilih kolom 'jenis_padi', 'harga_bibit', dan 'hasil_panen'
jenis_padi = dataset['jenis_padi']
harga_bibit = dataset['harga_bibit']
hasil_panen = dataset['hasil_panen']

# Gabungkan kolom-kolom ke dalam satu DataFrame baru
rekomendasi_df = pd.DataFrame({'jenis_padi': jenis_padi, 'harga_bibit': harga_bibit, 'hasil_panen': hasil_panen})

# Urutkan DataFrame berdasarkan hasil panen terbanyak dan harga terendah
rekomendasi_df = rekomendasi_df.sort_values(by=['hasil_panen', 'harga_bibit'], ascending=[False, True])

# Visualisasi hasil rekomendasi
fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar plot untuk 'hasil_panen' dan 'harga_bibit'
ax1.bar(rekomendasi_df.index - 0.2, rekomendasi_df['hasil_panen'], width=0.4, label='Hasil Panen', color='green', alpha=0.7)
ax1.bar(rekomendasi_df.index + 0.2, rekomendasi_df['harga_bibit'], width=0.4, label='Harga Bibit', color='blue', alpha=0.7)
ax1.set_title('Hasil Panen dan Harga Bibit berdasarkan Rekomendasi')
ax1.set_xlabel('jenis_padi')
ax1.set_ylabel('Harga Bibit')
ax1.legend()

# Bar plot untuk 'jenis_padi'
ax2 = ax1.twinx()  # Menggunakan sumbu y ganda untuk plot jenis_padi
ax2.bar(rekomendasi_df.index, rekomendasi_df['hasil_panen'], width=0.4, label='Jenis Padi', color='green', alpha=0.7)
ax2.set_ylabel('Ton')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset menggunakan Pandas
dataset = pd.read_csv('data - Sheet1.csv')

# Pilih kolom 'jenis_padi', 'harga_bibit', 'masa_tanam', dan 'jumlah_panen'
jenis_padi = dataset['jenis_padi']
harga_bibit = dataset['harga_bibit']
masa_tanam = dataset['masa_tanam']
jumlah_panen = dataset['jumlah_panen']

# Gabungkan kolom-kolom ke dalam satu DataFrame baru
rekomendasi_df = pd.DataFrame({'jenis_padi': jenis_padi, 'harga_bibit': harga_bibit, 'masa_tanam': masa_tanam, 'jumlah_panen': jumlah_panen})

# Urutkan DataFrame berdasarkan masa tanam (secara menaik)
rekomendasi_df = rekomendasi_df.sort_values(by='masa_tanam')

# Visualisasi hasil rekomendasi dalam bentuk bar plot
plt.figure(figsize=(12, 3))
plt.bar(rekomendasi_df['jenis_padi'], rekomendasi_df['masa_tanam'], color='blue')
plt.title('Rekomendasi Jenis Padi berdasarkan Masa Tanam (Lebih Cepat)')
plt.xlabel('Varietas Padi')
plt.ylabel('Masa Tanam (hari)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
